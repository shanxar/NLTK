## **USE A FILE FROM GUTENBERG CORPUS , PLOT THE MOST COMMON WORDS FROM IT**

```python
import nltk
nltk.download('all')
```

```python
from nltk.corpus import gutenberg
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
import matplotlib.pyplot as plt
import math

file_id=gutenberg.fileids()


#PREPROCESSING TEXT

#EXERCISE IS ABOUT WORDS FREQ, SO WE USE WORD TOKEN
#STEP 1 : TOKENIZING

words=gutenberg.raw(file_id[0])

word_tokens=word_tokenize(words.lower())

print(word_tokens)

#STEP 2:REMOVE STOP WORDS and NON-APLHANNUMERICS
stopwords_english=stopwords.words('english')
filtered_words=[word for word in word_tokens if word not in stopwords_english and word.isalnum()]


print(f"FILTERED WORDS :{filtered_words}")

#STEP 3: GET FREQ

word_freq=FreqDist(filtered_words)
top_5=(word_freq.most_common(5))
print("TOP 5 freq :",top_5)

#STEP 4: UNPACK

words,counts= zip(*top_5)

#STEP 5 : PLOT
y_upperlimt= math.ceil(max(counts)/100)*100
y_lowerlimit=math.floor(max(counts)/100)*100
colors=["red","blue"]
plt.figure(figsize=(10,20))
plt.bar(words,counts,color=colors)
plt.ylim(y_lowerlimit,y_upperlimt)
plt.yticks(range(y_lowerlimit,y_upperlimt+1,10))
plt.xlabel("WORDS",color="brown")
plt.ylabel("COUNTS",color="black")
plt.show()
```
